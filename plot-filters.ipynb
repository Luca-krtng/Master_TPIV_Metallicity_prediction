```python
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
import pyphot
from pyphot import UnitFilter
import pandas as pd
```

```python
root_dir = Path('/Users/lucakrattinger/LOCAL_FILES/synthetic/AMBRE')
root_filt = Path('/Users/lucakrattinger/LOCAL_FILES/synthetic/filter')

# Euclid 
# Import filters
VIS_filter = root_filt/'Euclid_VIS.vis.dat'
#Y_filter = root_filt/'Euclid_NISP.Y.dat'
J_filter = root_filt/'Euclid_NISP.J.dat'
CaHK_filter = root_filt/'CaHK_trans.txt'

# Load filter transmission data

wl_vis, tx_vis =np.loadtxt(VIS_filter, unpack=True)
wl_j, tx_j = np.loadtxt(J_filter, unpack=True)
wl_cahk, tx_cahk =  np.loadtxt(CaHK_filter, usecols=(0, 2), unpack=True)

# SDSS
# Import filters
SDSS_g = root_filt/'SLOAN_SDSS.g.dat'
SDSS_i = root_filt/'SLOAN_SDSS.i.dat'
SDSS_r = root_filt/'SLOAN_SDSS.r.dat'

# Load filter transmission data
wl_gS, tx_gS =np.loadtxt(SDSS_g, unpack=True)
wl_iS, tx_iS = np.loadtxt(SDSS_i, unpack=True)
wl_rS, tx_rS = np.loadtxt(SDSS_r, unpack=True)
    
# PS1
# Import filters
PS1_g = root_filt/'PAN-STARRS_PS1.g.dat'
PS1_i = root_filt/'PAN-STARRS_PS1.i.dat'
PS1_r = root_filt/'PAN-STARRS_PS1.r.dat'

# Load filter transmission data
wl_gP, tx_gP =np.loadtxt(PS1_g, unpack=True)
wl_iP, tx_iP = np.loadtxt(PS1_i, unpack=True)
wl_rP, tx_rP = np.loadtxt(PS1_r, unpack=True)
```

```python
fig, axes = plt.subplots(1,3, figsize=(15, 6))
# SDSS
ax0 = axes[0]
ax0.plot(wl_gS, tx_gS, label ='g band')
ax0.plot(wl_iS, tx_iS, label ='i band')
ax0.plot(wl_rS, tx_rS, label ='r band')
ax0.plot(wl_cahk, tx_cahk, label ='CaHK band')
ax0.set_xlabel('Wavelength (Å)',fontsize=15)
ax0.set_ylabel('Transmission',fontsize=15)
ax0.grid(True, linestyle='--', alpha=0.4)
ax0.set_title('SDSS',fontsize=18)
ax0.legend(loc='upper left')
ax0.set_ylim(0,0.55)

# PS1
ax1 = axes[1]
ax1.plot(wl_gP, tx_gP, label ='g band')
ax1.plot(wl_iP, tx_iP, label ='i band')
ax1.plot(wl_rP, tx_rP, label ='r band')
ax1.plot(wl_cahk, tx_cahk, label ='CaHK band')
ax1.set_xlabel('Wavelength (Å)',fontsize=15)
ax1.set_ylabel('Transmission',fontsize=15)
ax1.grid(True, linestyle='--', alpha=0.4)
ax1.set_title('Pan-STARRS 1',fontsize=18)
ax1.legend()

# Euclid
ax2 = axes[2]
ax2.plot(wl_vis, tx_vis, label ='VIS band')
ax2.plot(wl_j, tx_j, label ='J band')
ax2.plot(wl_cahk, tx_cahk, label ='CaHK band',color='red')
ax2.set_xlabel('Wavelength (Å)',fontsize=15)
ax2.set_ylabel('Transmissison',fontsize=15)
ax2.grid(True, linestyle='--', alpha=0.4)
ax2.set_title('Euclid',fontsize=18)
ax2.legend(loc='upper left')
ax2.set_ylim(0,1)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.savefig('Filters.svg')
plt.show()
```
